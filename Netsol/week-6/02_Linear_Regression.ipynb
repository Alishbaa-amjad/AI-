# Linear Regression Task 2

* Initialize X and Y points ranging from 1 to 100
* Initialize theta 1 value between 1-5 ( you can use like 0,1,3 something like that)
* Draw X and Y plot
* Draw cost func plot

---


```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Data generate karna (X aur Y)
X = np.linspace(0, 100, 100)
# Actual slope hum 2 rakhte hain (True slope = 2)
Y = 2 * X + np.random.normal(0, 10, 100) 

# 2. Mukhtalif Theta 1 values jinhe hum test karenge
thetas_to_test = [0, 1, 2, 3, 4, 5]
costs = []

# Cost (MSE) calculate karne ka function
def compute_cost(X, Y, theta1):
    predictions = theta1 * X
    return np.mean((predictions - Y) ** 2)

# --- PLOTTING ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Left Plot: Dataset and Hypothesis Lines
ax1.scatter(X, Y, color='blue', s=10, alpha=0.5, label='Actual Data')
colors = ['gray', 'orange', 'green', 'red', 'purple', 'brown']

for i, t1 in enumerate(thetas_to_test):
    hypo_line = t1 * X
    ax1.plot(X, hypo_line, label=f'Hypothesis: θ1={t1}', color=colors[i])

ax1.set_title('Dataset and Hypothesis Lines')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.legend()
ax1.grid(True)

# Right Plot: Cost Function Curve
theta_range = np.linspace(-1, 6, 100)
cost_curve = [compute_cost(X, Y, t) for t in theta_range]

ax2.plot(theta_range, cost_curve, color='black', alpha=0.7, label='Cost Function Curve')

# Red dots lagana tested theta values par
for t1 in thetas_to_test:
    current_cost = compute_cost(X, Y, t1)
    ax2.scatter(t1, current_cost, color='red', zorder=5)
    ax2.text(t1, current_cost + 500, f'θ1={t1}', ha='center')

ax2.set_title('Cost Function J(θ1)')
ax2.set_xlabel('Theta 1 (Slope)')
ax2.set_ylabel('Cost (Error)')
ax2.grid(True)

plt.tight_layout()
plt.show()
```
