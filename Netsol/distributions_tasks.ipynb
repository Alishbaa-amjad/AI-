```python
#Online order Success
# Ek e-commerce app hai jahan har order ya to success (1) hota hai ya fail (0) (payment issue).
import numpy as np
import matplotlib.pyplot as plt

# Assume success probability = 0.8
p = 0.8
orders = np.random.binomial(1, p, size=20)

print("Order Results:", orders)
print("Total Successful Orders:", np.sum(orders))

# Probabilities
p0 = np.mean(orders == 0)
p1 = np.mean(orders == 1)

# Plot
fig, ax = plt.subplots(figsize=(7,5))

bars = ax.bar([0,1], [p0, p1], color=['red', 'green'], edgecolor='black')

# Labels & Title
ax.set_title("Bernoulli Distribution (Online Orders)", fontsize=13, fontweight='bold')
ax.set_xticks([0,1])
ax.set_xticklabels(["Fail (0)", "Success (1)"])
ax.set_ylabel("Probability")

# Add percentage labels on bars
for bar, prob in zip(bars, [p0, p1]):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 0.02,
            f"{prob*100:.1f}%",
            ha='center', va='bottom', fontsize=10)

# Add grid
ax.yaxis.grid(True, linestyle='--', alpha=0.6)
ax.set_axisbelow(True)

# Remove extra borders
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# Har player coin ko 10 baar toss karta hain.
# Har toss ka outcome ya to Heads (1) ya Tails (0) hota hai.
# Hum dekhte hain har player ke 10 toss mein kitne Heads aaye.

from scipy.stats import binom

# 1. Define the Scenario Parameters
n_tosses = 10      # Number of coin tosses (n)
p_heads = 0.5      # Probability of getting heads (p = 50%)

# Generate the possible outcomes (0 to 10 heads)
outcomes = np.arange(0, n_tosses + 1)

# Calculate the binomial probability for each outcome
probabilities = binom.pmf(outcomes, n_tosses, p_heads)

# 2. Build the Visualization
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar chart using a nice contrasting color
bars = ax.bar(outcomes, probabilities, color='purple', edgecolor='black')

# 3. Polish and Annotate
ax.set_xticks(outcomes)
ax.set_xlabel('Number of Heads (Successes)',fontweight='bold')
ax.set_ylabel('Probability',fontweight='bold')
ax.set_title(f'Binomial Distribution: 10 Coin Tosses ($n={n_tosses}, p={p_heads}$)',fontweight='bold')

# Add the exact percentage text on top of each bar
for bar, prob in zip(bars, probabilities):
    height = bar.get_height()
    # Only add text if the probability is large enough to read easily (>0.1%)
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,f'{prob*100:.1f}%',ha='center',va='bottom')

# Add a subtle grid for easier reading
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# normal distribution
# Ek company mein employees ki salaries average ke around hoti hain
from scipy.stats import norm

# Parameters
mean_salary = 50000
std_salary = 8000
n_employees = 200

# Generate salaries
salaries = np.random.normal(mean_salary, std_salary, n_employees)

# X-axis for PDF curve
x = np.linspace(min(salaries), max(salaries), 200)
pdf = norm.pdf(x, mean_salary, std_salary)

# Plot
fig,ax= plt.subplots(figsize=(8,6))

# Histogram of salaries
plt.hist(salaries,bins=20,density=True, color='green',edgecolor='black',alpha=0.7,label='Employee Salaries')

# PDF curve
ax.plot(x, pdf, 'b-', linewidth=2, label='Normal PDF')

# Mean line
ax.axvline(mean_salary, color='red', linestyle='--', label='Mean Salary')

# Labels & Title
ax.set_xlabel('Salary',fontweight='bold')
ax.set_ylabel('Probability Density',fontweight='bold')
ax.set_title(f'Normal Distribution of Employee Salaries (mean={mean_salary}, std={std_salary})',fontweight='bold')

# Grid & Border cleanup
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Socho tum ek class teacher ho: Tum ne ek test liya.Lekin sirf 8 students class mein aaye
# Unke marks:[65, 70, 68, 72, 66, 74, 69, 71]
# Tumhe estimate karna hai:“Class ka average marks kitna hai?”
# Problem kya hai? Students bohat kam hain (small sample).Aur population ka actual standard deviation bhi nahi pata
# Is situation mein:❌ Normal Distribution use nahi hoti ✅ T-Distribution use hoti hai

from scipy.stats import t

# Sample data (marks of 8 students)
data = np.array([65, 70, 68, 72, 66, 74, 69, 71])  

# Mean and standard error
mean = np.mean(data)               #69.3
std = np.std(data, ddof=1)
n = len(data)
se = std / np.sqrt(n)

# X values
x = np.linspace(mean - 10, mean + 10, 100)  #59.3, 79.3

# t-distribution PDF
df = n - 1
pdf = t.pdf((x - mean)/se, df) #(x-mean)=(sample mean-population mean)

# Plot
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(x, pdf, color='blue', linewidth=2, label='t-Distribution')

# Mean line
ax.axvline(mean, color='red', linestyle='--', label='Sample Mean')

# Labels
ax.set_title("T-Distribution (Small Class Test Marks)", fontsize=13, fontweight='bold')
ax.set_xlabel("Marks")
ax.set_ylabel("Density")

# Styling
ax.yaxis.grid(True, linestyle='--', alpha=0.6)
ax.set_axisbelow(True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Socho tum ek juice brand test kar rahi ho: Tum poori city ka opinion nahi le sakti.
# Is liye tum sirf 5 logon se taste test karwati ho
# Unhone rating di (out of 10): [6, 7, 8, 7, 10]
# Ab tum soch rahi ho:“Kya ye juice logon ko pasand hai overall?”
# Problem kya hai? Sirf 5 log (bohat chhota data) --> Puri population ka idea nahi
# Is liye:❌ Normal distribution risky. ✅ T-distribution use hoti hai

# Sample data
data = np.array([6, 7, 8, 7, 10])

mean = np.mean(data) #7.6
std = np.std(data, ddof=1) 
n = len(data) #5

# X values
x = np.linspace(mean - 3, mean + 3, 100)  #4.6 , 10.6
fig, ax = plt.subplots(figsize=(8,5))

# t-distribution
df = n - 1   #4
y = t.pdf((x - mean)/(std/np.sqrt(n)), df)

# Plot
plt.plot(x, y)
plt.title("T-Distribution (Juice Taste Test)")
plt.xlabel("Rating")
plt.ylabel("Density")
plt.axvline(mean, color='red', linestyle='--')

# Styling
ax.yaxis.grid(True, linestyle='--', alpha=0.6)
ax.set_axisbelow(True)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
