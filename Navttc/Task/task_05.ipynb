```python
# Program to calculate square root using exponent operator

# Step 1: Define the number (or take user input)
num = 8 

# Step 2: Calculate square root (Power of 0.5 is equal to square root)
num_sqrt = num ** 0.5

# Step 3: Display result with 3 decimal places
print('The square root of %0.3f is %0.3f'%(num, num_sqrt))
```
