```python
# Program to add two numbers with user input

# Step 1: Taking input from the user
# input() returns a string, so we convert it to float for calculation
num1 = int(input('Enter first number: '))
num2 = int(input('Enter second number: '))

# Step 2: Add the two numbers
# float() ensures that decimal numbers are also handled correctly
sum = float(num1) + float(num2)

# Step 3: Display the result
# Using .format() to print numbers and their sum in a clear message
print('The sum of {0} and {1} is {2}'.format(num1, num2, sum))
```
